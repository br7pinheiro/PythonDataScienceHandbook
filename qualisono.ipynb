```python
import pandas as pd
import numpy as np
import unicodedata
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency, mannwhitneyu, kruskal, spearmanr
import statsmodels.formula.api as smf
from IPython.display import display

# --- 1. Carregar e limpar dados ---
df = pd.read_excel('/content/perfil_sono_medicina.xlsx')
df.columns = df.columns.str.strip().str.upper().str.replace(' ', '_')

def clean_text(s):
    if pd.isna(s): return ''
    s = str(s).strip().upper()
    return ''.join(c for c in unicodedata.normalize('NFKD', s) if not unicodedata.combining(c))

for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].apply(clean_text)

# --- 2. Converter e categorizar variáveis ---
df['IDADE_NUM'] = pd.to_numeric(df['IDADE'], errors='coerce')
df['FAIXA_IDADE'] = pd.Categorical(
    pd.cut(df['IDADE_NUM'], bins=[0,20,25,30,np.inf], labels=['<20','21-25','26-30','>30']),
    ordered=True)

df['MD_NUM'] = pd.to_numeric(df['MD'], errors='coerce')

df['HORADOSONO_CAT'] = pd.Categorical(
    df['HORADOSONO'], categories=['<22H','22-23H','23-00H','>00H'], ordered=True)

df['MEDIASONO_CAT'] = pd.Categorical(
    df['MEDIASONO'], categories=['<4H','4-5H','6-7H','>8H'], ordered=True)

df['QUALI_SONO_CAT'] = pd.Categorical(
    df['QUALI_SONO'], categories=['INSUFICIENTE','MAU','REGULAR','BOA','EXCELENTE'], ordered=True)

df['GENERO_BIN'] = df['GENERO']
df['TRABALHA_BIN'] = df['TRABALHA'].map({'S':'SIM','N':'NAO'})
df['CODICAO_PRE_BIN'] = df['CODICAO_PRE'].map({'S':'SIM','N':'NAO'})
df['SONO_SUFICIENTE_BIN'] = df['SONO_SUFICIENTE'].map({'S':'SIM','N':'NAO'})

df['MORADIA_CAT'] = pd.Categorical(
    df['MORADIA'], categories=['FAMILIA','SOZINHO','REPUBLICA'], ordered=True)

ord_vars = [
    'DIFICULDADEDORMIR','ACORDANOITE','CANSACO','DIFICULDADEACORDAR',
    'ESTRESSE_ACADEMICO','USODISPOSITIVO','BEBIDA','ATIVIDADE',
    'ATIVIDADEFISICA','DESCANSADO_MANHA','ESTUDA_NOITE','HORARIO_FIXO','PRATICA_RELAX'
]

freq_cats = ['NUNCA','RARAMENTE','ALGUMAS VEZES','FREQUENTEMENTE']
est_cats = ['NADA','POUCO','MODERADAMENTE','MUITO']

for var in ord_vars:
    df[var] = df[var].replace({'ALGUMAS':'ALGUMAS VEZES'})
    cats = est_cats if var == 'ESTRESSE_ACADEMICO' else freq_cats
    df[var + '_CAT'] = pd.Categorical(df[var], categories=cats, ordered=True)

# --- 3. Tabelas descritivas e gráficos por seção ---
sections = {
    'Sociais': ['FAIXA_IDADE','GENERO_BIN','MD_NUM','MORADIA_CAT','TRABALHA_BIN'],
    'Padrao_Sono': ['HORADOSONO_CAT','MEDIASONO_CAT','DIFICULDADEDORMIR_CAT',
                    'ACORDANOITE_CAT','CANSACO_CAT','DIFICULDADEACORDAR_CAT'],
    'Fatores_Sono': ['ESTRESSE_ACADEMICO_CAT','USODISPOSITIVO_CAT',
                     'BEBIDA_CAT','ATIVIDADE_CAT','CODICAO_PRE_BIN'],
    'Qualidade_Sono': ['ATIVIDADEFISICA_CAT','QUALI_SONO_CAT',
                       'DESCANSADO_MANHA_CAT','ESTUDA_NOITE_CAT','PIORA_MEDICINA'],
    'Habitos_Sono': ['HORARIO_FIXO_CAT','PRATICA_RELAX_CAT','SONO_SUFICIENTE_BIN']
}

for sec, vars in sections.items():
    rows = []
    for var in vars:
        freq = df[var].value_counts(dropna=False).rename('Abs_Freq')
        rel = (freq / len(df)).round(3).rename('Rel_Freq')
        tbl = pd.concat([freq, rel], axis=1).reset_index().rename(columns={'index':'Categoria'})
        tbl.insert(0, 'Variavel', var)
        rows.append(tbl)
        plt.figure()
        freq.sort_index().plot(kind='bar')
        plt.title(var)
        plt.xlabel(var)
        plt.ylabel('Frequência')
        plt.tight_layout()
        plt.show()
    sec_df = pd.concat(rows, ignore_index=True)
    display(sec_df)

# --- 4. Qui-quadrado vs QUALI_SONO_CAT ---
chi2_res = []
for var in sum(sections.values(), []):
    if var != 'QUALI_SONO_CAT':
        ct = pd.crosstab(df[var], df['QUALI_SONO_CAT'])
        p = np.nan
        if ct.shape[0] > 1 and ct.shape[1] > 1:
            _, p, _, _ = chi2_contingency(ct)
        chi2_res.append({'Variavel': var, 'p_valor': round(p,4)})
chi2_df = pd.DataFrame(chi2_res)
display(chi2_df)

# --- 5. Mann-Whitney vs MEDIASONO_CAT ---
mw_res = []
for var in ['TRABALHA_BIN','GENERO_BIN','CODICAO_PRE_BIN','SONO_SUFICIENTE_BIN']:
    codes = df['MEDIASONO_CAT'].cat.codes
    grp0 = codes[df[var]=='NAO']
    grp1 = codes[df[var]=='SIM']
    p = np.nan
    if len(grp0) > 0 and len(grp1) > 0:
        _, p = mannwhitneyu(grp0, grp1)
    mw_res.append({'Variavel': var, 'p_valor': round(p,4)})
mw_df = pd.DataFrame(mw_res)
display(mw_df)

# --- 6. Kruskal-Wallis vs MEDIASONO_CAT ---
kw_res = []
for var in ['MORADIA_CAT','FAIXA_IDADE','HORADOSONO_CAT','ESTRESSE_ACADEMICO_CAT']:
    groups = [df.loc[df[var]==cat,'MEDIASONO_CAT'].cat.codes for cat in df[var].cat.categories]
    groups = [g for g in groups if len(g) > 0]
    p = np.nan
    if len(groups) > 1:
        _, p = kruskal(*groups)
    kw_res.append({'Variavel': var, 'p_valor': round(p,4)})
kw_df = pd.DataFrame(kw_res)
display(kw_df)

# --- 7. Spearman entre variáveis ordinais ---
spe_res = []
pairs = [
    ('FAIXA_IDADE','MEDIASONO_CAT'),
    ('DIFICULDADEDORMIR_CAT','MEDIASONO_CAT'),
    ('CANSACO_CAT','MEDIASONO_CAT'),
    ('HORADOSONO_CAT','MEDIASONO_CAT')
]
for x_var, y_var in pairs:
    x = df[x_var].cat.codes
    y = df[y_var].cat.codes
    rho, p = spearmanr(x, y, nan_policy='omit')
    spe_res.append({'X': x_var, 'Y': y_var, 'rho': round(rho,3), 'p_valor': round(p,4)})
spe_df = pd.DataFrame(spe_res)
display(spe_df)

# --- 8. Regressão logística exploratória ---
df['QUALI_HIGH'] = df['QUALI_SONO_CAT'].isin(['BOA','EXCELENTE']).astype(int)
model = smf.logit('QUALI_HIGH ~ C(GENERO_BIN)', data=df).fit(disp=False)
display(model.summary())
```
